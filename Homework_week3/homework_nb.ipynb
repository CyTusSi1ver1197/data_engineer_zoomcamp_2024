```python
#download data
months = ["01", "02", "03", "04", "05","06","07", "08", "09", "10", "11", "12"]
data_df = []
full_taxi_df = pd.DataFrame()

for month in months:
    # Online URL to get data
    url = "https://d37ci6vzurychx.cloudfront.net/trip-data/green_tripdata_2022-" + month + ".parquet"

    # local url if already downloaded (on cmd input python -m http.server 
    # open another cmd then input ipconfig to find port)
    # url = "http://172.25.240.1:8000/"

    full_url = url
    response = rqts.get(full_url)
    
    
    ans_df = pd.read_parquet(full_url, engine= 'auto')
    data_df.append(ans_df)

full_taxi_df = pd.concat(data_df, axis = 0, ignore_index = True)    
```
